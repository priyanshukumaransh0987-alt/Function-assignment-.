##1. Difference between a Function and a Method in Python

Function: A block of code that performs a specific task, defined using def. It is not bound to any object.

Method: A function that belongs to an object (usually called using object.method()).

Example:

```python
# Function
def add(a, b):
    return a + b

print(add(2, 3))   # Output: 5

# Method (belongs to string object)
text = "hello"
print(text.upper())  # Output: HELLO
```

##2. Function Arguments and Parameters

Parameter: Variable defined in the function header.

Argument: Actual value passed when calling the function.

Example:

```python
def greet(name):   # 'name' is a parameter
    print("Hello", name)

greet("Priya")     # "Priya" is an argument
```

## 3. Ways to Define and Call a Function

Normal function using def.

Lambda function.

Built-in functions.

 Example:

```python
# Normal
def square(x):
    return x*x
print(square(5))

# Lambda
sq = lambda x: x*x
print(sq(5))
```

##4. Purpose of return Statement

Used to send a result back from a function.

Without return, function outputs None.

Example:

```python
def add(a, b):
    return a + b

result = add(10, 5)
print(result)  # Output: 15
```

##5. Iterators vs Iterables

Iterable: An object that can return an iterator (list, tuple, string).

Iterator: An object with __next__() method to fetch values one at a time.

Example:



```python
numbers = [1, 2, 3]
it = iter(numbers)   # create iterator

print(next(it))  # 1
print(next(it))  # 2
```

##6. Generators in Python

Special functions that use yield instead of return.

They produce values lazily (one at a time).

 Example:



```python
def count_up_to(n):
    for i in range(1, n+1):
        yield i

for num in count_up_to(5):
    print(num)
```

## 7. Advantages of Generators

Memory efficient (does not store all values at once).

Faster for large datasets.

Useful for infinite sequences.

Example:



```python
gen = (x*x for x in range(5))  # generator expression
print(list(gen))  # [0, 1, 4, 9, 16]
```

##8. Lambda Functions

Anonymous (nameless) functions defined with lambda.

Used when function is short and used only once.

Example:



```python
square = lambda x: x*x
print(square(6))  # Output: 36
```

## 9. Purpose of map()

Applies a function to every element of an iterable.

Returns a new map object (convert to list to see results).

Example:




```python
nums = [1, 2, 3, 4]
squares = list(map(lambda x: x*x, nums))
print(squares)  # [1, 4, 9, 16]
```

##10. Difference between map(), reduce(), and filter()

map() → Transforms each element.

filter() → Selects elements based on a condition.

reduce() → Combines elements into a single result (from functools).

Example:


```python

from functools import reduce

nums = [1, 2, 3, 4]

print(list(map(lambda x: x*2, nums)))        # [2, 4, 6, 8]
print(list(filter(lambda x: x%2==0, nums)))  # [2, 4]
print(reduce(lambda a,b: a+b, nums))         # 10
```

```python
# Q1- Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list?
def sum_even(a):
  n= 0
  for i in a:
    if i % 2== 0:
      n = n+i
  return n
```

```python
print("Sum of even numbers: ", sum_even([1,2,3,5,6,77,]))
```

```python
#Q4. Write a Python function that checks if a given number is prime or not from 1 to 200?
def is_prime(num):
    if num < 2 or num > 200:   # restrict to 1–200
        return False
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            return False
    return True
```

```python
print(is_prime(7))    # True
print(is_prime(10))   # False
print(is_prime(199))  # True
print(is_prime(201))  # False (outside 1–200)
```

```python
#7. Implement a generator function that reads a file line by line and yields each line as a string?
def read_file_line_by_line(filename):
    file = open(filename, "r")
    for line in file:
        yield line.strip()
    file.close()

for line in read_file_line_by_line("samples.txt"):
    print(line)

```

```python
#Q8. Use a lambda function in Python to sort a list of tuples based on the second element of each tuple.

# Sample list of tuples
data = [(1, 3), (4, 1), (2, 5), (7, 2)]

# Sorting using lambda (based on 2nd element of each tuple)
sorted_data = sorted(data, key=lambda x: x[1])

print("Original list:", data)
print("Sorted list:", sorted_data)
```

```python
#Q9. Write a Python program that uses map() to convert a list of temperatures from Celsius to Fahrenheit.

# Sample list of temperatures in Celsius
celsius = [0, 20, 37, 100]

# Convert to Fahrenheit using map() and lambda
fahrenheit = list(map(lambda c: (c * 9/5) + 32, celsius))

print("Celsius temperatures:", celsius)
print("Fahrenheit temperatures:", fahrenheit)
```

```python
#Q10. Create a Python program that uses filter() to remove all the vowels from a given string.

# Input string
text = "Programming is fun"

# Define vowels
vowels = "aeiouAEIOU"

# Use filter() with lambda to remove vowels
result = "".join(filter(lambda ch: ch not in vowels, text))

print("Original string:", text)
print("Without vowels:", result)
```
